```python
import json
import time

from tqdm.notebook import trange, tqdm
```

```python
with open('ERAI Dataset_with_translation/ERAI_Dataset_unsupervised_with_translation.json') as fp:
    data = json.load(fp)
```

```python
data[88]
```

```python
def zh_len_scorer(text, tokenizer):
    outputs = tokenizer(text)
    return len(outputs['input_ids'])
```

```python
scores = {} # index:(label, score)
```

```python
for d in tqdm(data):
    text = d['post_rationale']
    scores[d['index']] = zh_len_scorer(text, tokenizer=tokenizer)
```

```python
with open('BuildingLexicon_code/scores/bert-zh-tokenize-len.json', 'w') as f:
    json.dump(scores, f)
```

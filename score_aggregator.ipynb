```python
files = os.listdir('.')
```

```python
with open('bert-zh-tokenize-len.json', 'r') as f:
    len_score = json.load(f)
```

```python
with open('FinBERT-SpecificFLS.json', 'r') as f:
    fls_score = json.load(f)
```

```python
with open('FinBERT-Non-specificFLS.json', 'r') as f:
    ns_fls_score = json.load(f)
```

```python
with open('FinBERT-Not-FLS.json', 'r') as f:
    not_fls_score = json.load(f)
```

```python
with open('FinProLex.json', 'r') as f:
    finprolex_score = json.load(f)
```

```python
with open('FinWord.json', 'r') as f:
    finword_score = json.load(f)
```

# MPP = len + FLS + 0.5 NON-FLS - NOT-FLS + FinProLex + |FinWord>0|

```python
keys = list(range(210))
```

```python
mpp_scores = {}
```

```python
for k in keys:
    mpp_scores[k] = 0.0
    mpp_scores[k] += len_score[str(k)]
    mpp_scores[k] += fls_score[str(k)]
    mpp_scores[k] += 0.5*ns_fls_score[str(k)]
    mpp_scores[k] -= not_fls_score[str(k)]
    mpp_scores[k] += finprolex_score[str(k)]
    if finword_score[str(k)] > 0:
        mpp_scores[k] += finword_score[str(k)]
```

```python
mpp_scores
```

```python
mpp = sorted(mpp_scores,key=mpp_scores.get,reverse=True)
```

```python
with open('ERAI_unsupervised_MPP_PromptShots_0.json','w') as f:
    json.dump(mpp, f)
```

# ML = len + FLS + 0.5 NON-FLS - NOT-FLS + FinProLex + |FinWord<0|

```python
keys = list(range(210))
```

```python
ml_scores = {}
```

```python
for k in keys:
    ml_scores[k] = 0.0
    ml_scores[k] += len_score[str(k)]
    ml_scores[k] += fls_score[str(k)]
    ml_scores[k] += 0.5*ns_fls_score[str(k)]
    ml_scores[k] -= not_fls_score[str(k)]
    ml_scores[k] += finprolex_score[str(k)]
    if finword_score[str(k)] < 0:
        ml_scores[k] += -finword_score[str(k)]
```

```python
ml = sorted(ml_scores,key=ml_scores.get,reverse=False)
```

```python
with open('ERAI_unsupervised_ML_PromptShots_0.json','w') as f:
    json.dump(ml, f)
```

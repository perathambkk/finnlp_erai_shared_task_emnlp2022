```python
import json
import time

from tqdm.notebook import trange, tqdm
```

```python
with open('BuildingLexicon_code/FinProLex.json','r') as f:
    finprolex = json.load(f)
```

```python
with open('ERAI Dataset_with_translation/ERAI_Dataset_unsupervised_with_translation.json') as fp:
    data = json.load(fp)
```

```python
def finprolex_scorer(text, lexicon):
    score = 0.0
    for l in lexicon:
        if l['token'] in text:
            score += l['expertise_score']
    return score
```

```python
scores = {} # index:(label, score)
```

```python
for d in tqdm(data):
    text = d['post_rationale']
    scores[d['index']] = finprolex_scorer(text, lexicon=finprolex)
```

```python
with open('BuildingLexicon_code/scores/FinProLex.json', 'w') as f:
    json.dump(scores, f)
```

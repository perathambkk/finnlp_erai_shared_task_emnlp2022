```python
with open('bert-zh-tokenize-len.json', 'r') as f:
    len_score = json.load(f)
```

```python
with open('FinBERT-SpecificFLS.json', 'r') as f:
    fls_score = json.load(f)
```

```python
with open('FinBERT-Non-specificFLS.json', 'r') as f:
    ns_fls_score = json.load(f)
```

```python
with open('FinBERT-Not-FLS.json', 'r') as f:
    not_fls_score = json.load(f)
```

```python
with open('ERAI-finprolex-res.json', 'r') as f:
    finprolex_score = json.load(f)
```

```python
with open('ERAI-finwords-res.json', 'r') as f:
    finword_score = json.load(f)
```

```python
with open('ERAI-verbs.json', 'r') as f:
    verb_score = json.load(f)
```

```python
with open('ERAI-nouns.json', 'r') as f:
    noun_score = json.load(f)
```

```python
with open('ERAI-cards.json', 'r') as f:
    card_score = json.load(f)
```

```python
with open('FinBERT-tone_Negative.json', 'r') as f:
    tone_neg_score = json.load(f)
```

```python
with open('FinBERT-tone_Positive.json', 'r') as f:
    tone_pos_score = json.load(f)
```

```python
with open('ERAI-res.json', 'r') as f:
    tone_lex_score = json.load(f)
```

# MPP = (len + FinProLex) + (FLS + 0.5 NON-FLS - NOT-FLS) + (pos_tone - neg_tone + lex) + (nouns + cards - verbs) + |FinWord>0|

```python
keys = list(range(210))
```

```python
mpp_scores = {}
```

```python
for k in keys:
    mpp_scores[k] = 0.0
    mpp_scores[k] += len_score[str(k)]
    mpp_scores[k] += finprolex_score[k]
    
    mpp_scores[k] += fls_score[str(k)]
    mpp_scores[k] += 0.5*ns_fls_score[str(k)]
    mpp_scores[k] -= not_fls_score[str(k)]
    
    mpp_scores[k] += tone_pos_score[str(k)]
    mpp_scores[k] -= tone_neg_score[str(k)]
    mpp_scores[k] += tone_lex_score[k]
    
    mpp_scores[k] += noun_score[k]
    mpp_scores[k] += card_score[k]
    mpp_scores[k] -= verb_score[k]

    if finword_score[k] > 0:
        mpp_scores[k] += finword_score[k]
```

```python
mpp_scores
```

```python
mpp = sorted(mpp_scores,key=mpp_scores.get,reverse=True)
```

```python
with open('ERAI_unsupervised_MPP_PromptShots_1.json','w') as f:
    json.dump(mpp, f)
```

# MPP = (len + FinProLex) + (FLS + 0.5 NON-FLS - NOT-FLS) + (pos_tone - neg_tone + lex) + (nouns + cards - verbs) + |FinWord>0|

```python
keys = list(range(210))
```

```python
ml_scores = {}
```

```python
for k in keys:
    ml_scores[k] = 0.0
    ml_scores[k] += len_score[str(k)]
    ml_scores[k] += finprolex_score[k]
    
    ml_scores[k] += fls_score[str(k)]
    ml_scores[k] += 0.5*ns_fls_score[str(k)]
    ml_scores[k] -= not_fls_score[str(k)]
    
    ml_scores[k] += tone_pos_score[str(k)]
    ml_scores[k] -= tone_neg_score[str(k)]
    ml_scores[k] += tone_lex_score[k]
    
    ml_scores[k] += noun_score[k]
    ml_scores[k] += card_score[k]
    ml_scores[k] -= verb_score[k]

    if finword_score[k] < 0:
        ml_scores[k] += -finword_score[k]
```

```python
finword_score
```

```python
ml = sorted(ml_scores,key=ml_scores.get,reverse=False)
```

```python
with open('ERAI_unsupervised_ML_PromptShots_1.json','w') as f:
    json.dump(ml, f)
```
